```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge, LogisticRegression
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
x_train, x_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=22, test_size=0.2)
```

```python
fig = plt.figure(figsize=(8, 6), dpi=100)
plt.xticks(range(cancer.data.shape[1]), labels=cancer.feature_names, rotation=90)
list = [0.1, 1, 10, 100]
for i in list:
    model = LogisticRegression(C=i)
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)
    label = "C=" + str(i)
    plt.plot(model.coef_.T, label=label)
    plt.legend()
```
