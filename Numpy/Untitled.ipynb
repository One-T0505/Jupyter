## numpy.unique()

用于给传入的张量去重，去除重复值，所有的值都只保留一个。 语法格式： numpy.unique(arr, return_index, return_inverse, return_counts)
> 参数说明：
 - arr : 传入的数组，若是多维数组，则以一维数组形式展开
 - return_index :    如果为True，则返回新数组元素在原数组中的位置（索引）
 - return_inverse :   如果为True，则返回原数组元素在新数组中的位置（索引）
 - return_counts :   如果为True，则返回去重后的数组元素在原数组中出现的次数

## numpy.argwhere()
函数中可以添加逻辑表达式，返回符合表达式的索引；若是多维数组则返回行，列索引组成的索引坐标

```python
a = np.array([10, 9, 9, 9, 20, 5, 4, 2, 8 ,3, 3, 11, 5, 8, 4])
np.argwhere((a<15) & (a>5))
```

```python
a[(a<15) & (a>5)]
```
