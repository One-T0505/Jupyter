## 绘图操作

```python
%matplotlib inline
import pandas as pd
import numpy as np
```

##### 绘制 Series

```python
s = pd.Series(np.random.randn(10), index=np.arange(0, 100, 10))
s.plot()
```

##### 绘制 DataFrame

DataFrame 绘图 默认会把每一列属性画一条线

```python
df = pd.DataFrame(np.random.rand(7,3), columns=['k1' ,'k2', 'k3'])
df.plot()
```
