## 柱状图

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = range(5)
y = np.random.randint(20, 50, 5)
```

```python
plt.title('bar')
plt.grid()
plt.bar(x, y)
```

###### 设置底部起始位置

```python
plt.title('bar')
plt.grid()
plt.bar(x, y, bottom=[5, 2, 10, 0, 6])
```

###### 设置bar 的颜色
 - color 属性可以为不同的 bar 设置不同的颜色
 - facecolor 只能设置一个颜色，整个柱状图只能是一个颜色

```python
plt.title('bar')
plt.grid()
plt.bar(x, y, color=['r', 'g', 'b', 'y', 'black'])
```

```python
plt.title('bar')
plt.grid()
plt.bar(x, y, color='black')
```

###### 其他样式设置

```python
plt.title('bar')
plt.bar(x, y, color='black', edgecolor='red', linestyle='--', linewidth=1.5)
```

### 同位置多柱状图

绘制该奖牌榜前五名

```python
countries = ['挪威', '德国', '中国', '美国', '瑞典']
gold = [16, 12, 9, 8, 8]
silver = [8, 10, 4, 10, 5]
bronze = [37, 27, 15, 25, 18]
width = 0.2
x = np.arange(len(countries))
plt.bar(x , gold, color='gold', width=0.2)
plt.bar(x+width , silver, color='silver', width=0.2)
plt.bar(x+2*width , bronze, color='saddlebrown', width=0.2)
```

再用 plt.xticks() 方法去调整X轴刻度

观察到坐标值在金牌下面，想让刻度移动到银色下面，只需将整个刻度往右移动一个 width，再将 labels 设置为 国家名

```python
countries = ['挪威', '德国', '中国', '美国', '瑞典']
gold = [16, 12, 9, 8, 8]
silver = [8, 10, 4, 10, 5]
bronze = [37, 27, 15, 25, 18]
width = 0.2
x = np.arange(len(countries))
# 设置显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.bar(x , gold, color='gold', width=0.2)
plt.bar(x+width , silver, color='silver', width=0.2)
plt.bar(x+2*width , bronze, color='saddlebrown', width=0.2)
plt.xticks(x+width, countries)
```

##### 优化

```python
plt.rcParams['figure.dpi'] = 100
plt.ylim(0, 40)
countries = ['挪威', '德国', '中国', '美国', '瑞典']
gold = [16, 12, 9, 8, 8]
silver = [8, 10, 4, 10, 5]
bronze = [37, 27, 15, 25, 18]
width = 0.2
x = np.arange(len(countries))
x_gold = x
x_silver = x + width
x_bronze = x + 2*width
# 设置显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.bar(x_gold , gold, color='gold', width=0.2, label='金牌')
plt.bar(x_silver , silver, color='silver', width=0.2, label='银牌')
plt.bar(x_bronze , bronze, color='saddlebrown', width=0.2, label='铜牌')
plt.xticks(x+width, countries)
plt.title('北京冬奥会奖牌榜')
plt.ylabel('奖牌数')
plt.legend(loc='upper right')

for i in np.arange(len(countries)):
    plt.text(x_gold[i], gold[i]+0.4, gold[i], ha='center')
    plt.text(x_silver[i], silver[i]+0.4, silver[i], ha='center')
    plt.text(x_bronze[i], bronze[i]+0.4, bronze[i], ha='center')
```

## 堆叠图

```python

plt.bar(countries, gold, color='gold', width=0.3)
plt.bar(countries, silver, color='silver', bottom=gold, width=0.3)
plt.bar(countries, bronze, color='saddlebrown', bottom=np.array(gold)+np.array(silver), width=0.3)
```
