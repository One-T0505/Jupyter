## 中文字体设置

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5, 5)
y = np.sin(x)
```

```python
plt.title('y等于sinx')
plt.plot(x, y)
```

matplotlib 中默认是无法处理中文字体的 应该如何处理？

解决办法：①设置字体样式及正常显示   ②但是中文里负号不能显示 在设置符号的显示

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('y等于sinx')
plt.plot(x, y)
```
